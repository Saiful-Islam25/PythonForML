## Arithmetic Operation

```python
import pandas as pd
var=pd.DataFrame({"A":[1,2,3,4,5],"B":[6,7,8,9,10]})
var
```

```python
var["C"]=var["A"]+var["B"]
var
```

```python
var["C"]=var["A"]-var["B"]
var
```

```python
var["C"]=var["A"]*var["B"]
var
```

```python
var["C"]=var["A"]/var["B"]
var
```

```python
var["C"]=var["A"]%var["B"]
var
```

```python
var["Python"]=var["A"]<=3
var
```

```python
var["Python"]=var["A"]<=3
var["Python"]=var["B"]>=7
var
```
